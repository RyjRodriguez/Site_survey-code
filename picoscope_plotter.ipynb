```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
filename = r"C:\Users\rrodr\Google Drive\Temp Work Folder\Enam lab site data\MFL0124_20200911-0001_upperright\MFL0124_20200911-0001_upperright_4.csv"
df = pd.read_csv(filename, skiprows=3, names=['Hz', 'dBv'])
```

```python
#convert dBv to V_m
df['v_m'] = df['dBv'].apply(lambda x: 10.00 ** (x /20.00))

#convert V_m to acc w/ 10 volt = 1g
df['acc'] = df['v_m'].apply(lambda x: x/10 * 9.81)

#convert acc to displacement (um)
df['dis'] = (0.5 * df['acc'] * (1/df['Hz']) ** 2) * 10**6
df['log'] = np.log10(df['dis'])

#convert acc/dis to velocity (in/sec)
df['vel'] = np.sqrt(2 * df['acc'] * df['dis']) * (1 / 25400.00)
```

```python
# Hz = list(df['Hz'])
# dis = list(df['dis'])
```

```python
#Make plot

plt.plot(list(df['Hz']), list(df['vel']), linewidth=0.5)
plt.xlabel('Hz')
plt.ylabel('vel, in/s')
plt.yscale('log')
plt.ylim([10**(-10), 10**(-4)])
plt.grid(linestyle='-')
plt.axvspan(40, 60, alpha=0.5, color='red')
plt.title('Vibration Spectrum from Lab XXXX')
# plt.savefig(filename + 'velocity.png', dpi=200)
plt.save()
plt.show()
```
